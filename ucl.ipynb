```python
col=[]
for i in df.columns:
    if df[i].dtype=="object":
        col.append(i)
print(col)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()

for i in col:
    df[i]=le.fit_transform(df[i])
```

```python
cor=df.corr()
plt.figure(figsize=(10,10))
print(sns.heatmap(cor,annot=True))
print(cor["income"].sort_values(ascending=False))
```

```python
from imblearn.over_sampling import SMOTE
smote=SMOTE()

x=df.drop("income",axis=1)
y=df["income"]


x,y=smote.fit_resample(x,y)
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
df=sc.fit_transform(df)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score
k_value=range(1,26)
score=[]

for k in k_value:
    knn=KNeighborsClassifier(k)
    scors=cross_val_score(knn,x,y,cv=10,scoring="accuracy")
    score.append(scors.mean())
```

```python
plt.plot(k_value,score,marker="o")
```

```python
knn=KNeighborsClassifier(n_neighbors=26)


knn.fit(x_train,y_train)
y_pred_train=knn.predict(x_train)
train_acc=accuracy_score(y_train,y_pred_train)


knn.fit(x_train,y_train)
y_pred_test=knn.predict(x_test)
test_acc=accuracy_score(y_test,y_pred_test)


print("train_accuracy=",round(train_acc,2))
print("test_accuracy=",round(test_acc,2))
```

```python
from sklearn.metrics import roc_curve, auc

y_scores=knn.predict_proba(x_test)[:,1]
fpr,tpr,thresholds=roc_curve(y_test,y_scores)
roc_auc=auc(fpr,tpr)
```

```python
plt.figure()
plt.plot(fpr,tpr,color="navy")
plt.xlabel("false positive rate")
plt.ylabel("true positive rate")
```
